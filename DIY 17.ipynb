# Data Visualization using Matplotlib
Load the “Netflix_titles.csv”data  into  a  DataFrame  and  plot  the  following insights using the Matplotliblibrary:1.How much of the Netflix content areTV showsor Movies?2.In which country is Netflix the most popular?

# 1.How much of the Netflix content areTV shows or Movies?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv("netflix_titles.csv")
df
```

```python
#obtaining the unique elements from the type column and convert as list
label=list(set(df.type))

# creating an empty list
count=[]

#appending the dataframe of the 
for i in label:
    count.append(len(df[df.type==i]))


minim=count.index(min(count))
exp=np.zeros(len(count))

# setting the mimimum distance between the partition of pie 
exp[minim]=0.1

#plotting a pie chart with the labels and the partition percentage is displayed autopct
# dividing the partition in explode 
plt.pie(count,labels=label,autopct='%0.2f',explode=exp)

#Getting a perfect circle 
plt.axis('equal')
plt.title("Type of Netflix Content",color='g')
plt.show()
    
    
```

# 2.In which country is Netflix the most popular?

```python
#Counting the country  and create a dataframe with reset index 
dcount= df['country'].value_counts().reset_index()
dcount= dcount[dcount['country'] / dcount['country'].sum() > 0.01]
dcount
```

```python
plt.figure(figsize=(12,10))
plt.pie(dcount['country'],labels=dcount['index'])
plt.title('Netflix Viewers across the Country',color='b',fontsize=20)
plt.axis('equal')
plt.show()
```

# Q2. Problem Statement:Data Visualization using Seaborn
Load the “starbucks_drinkMenu_expanded.csv”data into a DataFrameand find out which “Beverage_category”has the highest calories;plot the insights using a bar plot with the help of theSeabornlibrary.

```python
import pandas as pd

#importing the seaborn 
import seaborn as sb
import matplotlib.pyplot as plt

#open the starbucks_drinkMenu_expanded csv file
df=pd.read_csv("starbucks_drinkMenu_expanded.csv")
df
```

```python
#Set the figure size of the graph
plt.figure(figsize=(10,8))

# drawing the barplot with the Calories as X axis and Beverage_category as Y axis 
# the color palette  as hue, saturation, and luminance
sb.barplot(x = "Calories", y = "Beverage_category", palette = 'hls', data = df)
plt.show()
```
